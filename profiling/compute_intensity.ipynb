```python
m = 128
n = 128
k = 64
```

```python
def calc_ai(m, n, k, a, b):
    src_ai = (2 * m * n * k) / (m * k + k * n + m * n)
    dst_ai = (2 * ((1 / a) * m) * (b * n) * k) / ((1 / a) * m * k + k * (b * n) + (1 / a) * m * (b * n))
    # dst_ai = (2 * m * n * k) / ((1 / a) * m * k + k * n + (1 / a) * m * n)
    return src_ai, dst_ai
```

```python
a = 4
b = 7 / 4
src_ai, dst_ai = calc_ai(m, n, k, a, b)
print(f"src_ai: {src_ai}, dst_ai: {dst_ai}")
print(f"m / a: {m / a}, n * b: {n * b}")
```
